2. Analyze Matrix A
Using NumPy, compute:
● Shape of A
● Determinant
● Rank
● Eigenvalues
● Inverse (only if determinant ≠ 0)

```python

d1 = 0
d2 = 2
A = np.array([[d1+2, d2+1],[2*d1, d2+2]])

size_A=np.shape(A)
det_A=np.linalg.det(A)
rank_A=np.linalg.matrix_rank(A)
eigenvalues=np.linalg.eig(A)
if np.linalg.det(A) != 0:
    print("Inverse:\n", np.linalg.inv(A))
else:
    print("Inverse: Not possible (determinant is 0)")

#display results

print("Shape:", size_A)
print("Determinant:", np.linalg.det(A))
print("Rank:", np.linalg.matrix_rank(A))
print("Eigenvalues:", np.linalg.eigvals(A))
```

3. Value Change Experiment
Create a new matrix B by changing exactly one value in A:
● Add +1 to one element of your choice
● You must clearly state which element you changed


```python
A = np.array([[2, 3],[0, 4]])

B = A.copy()
B[1,0] = B[1,0] + 1

print("\nMatrix B (changed A[1,0] by +1):")
print(B)
```

1. Determinant change and why:

Ans: The determinant changed because one of the values in the matrix was modified. Since the determinant depends on all the elements of the matrix, even a small change in one entry can increase or decrease its value.

2. Did the rank change? Explain.

Ans: No, the rank usually stays the same if the rows and columns are still independent after the change. Since only a value was adjusted (and no row became a copy or multiple of another), the rank likely did not change.

3. How did the eigenvalues respond to the value change?

Ans: The eigenvalues also changed because they depend on the matrix entries. When one number in the matrix was changed, the eigenvalues shifted slightly, reflecting the new structure of the matrix.

4. Is B easier or harder to invert than A? Why?
